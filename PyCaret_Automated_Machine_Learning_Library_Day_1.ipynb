# Pycaret - Handling Missing Values ( NaN Values )

```
from pycaret.datasets import get_data
hepatits=get_data('hepatitis')
```

```
# In pycaret if we dont apply any technique to handle NAN values , Bydefault it will replace the NAN values with Mean.
# By default numeric_imputation='mean'

from pycaret.classification import *
# Data preprocessing is done using setup() in pycaret
clf=setup(data=hepatits,target='Class')
```

```
# Getting other data set
pokemon=get_data('pokemon')
```

```
# If we dont want few features from the data set we can remove it by using 'ignore_features' in pycaret
setup(data=pokemon,target="Legendary",ignore_features=["#","Name"])
```

```
hepatitis=get_data("hepatitis")
```

```
# changing some feature from numerical to categorical
# Suppose we want to change Age which is numeric into categorical feature we can use 'categorical_feature()' in pycaret

setup(data=hepatitis,target='Class',categorical_features=["AGE"])
```

```
# By default for categorical features pycaret applies one hot encoding 'high_cardinality_feature' 
# Lets see ths using pokemon dataset itself where type1,type2 are categorical features

setup(data=pokemon,target="Legendary")
```

```
# Ordinal encoding (assigning ranks)
employee=get_data('employee')
```

```
employee['salary'].value_counts()
```

```
# lets assign rank to salary feature in the dataset

setup(data=employee,target='left',ordinal_features={'salary':['low','medium','high']})
```

```
# cardinality 

income=get_data('income')
```

```
income['native-country'].unique()
income['native-country'].value_counts()
```

```
# we are taking native-country as high cardinality because it has more unique values
setup(data=income,target='income >50K',high_cardinality_features=['native-country'])
```
